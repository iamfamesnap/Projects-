```python
# Formula to calculate recovery:
# Recovery = [(C * (F - T)) / (F * (C - T))] * 100
# where:
# C = concentration of the target component (gold) in the concentrate right after flotation (or cleaning);
# F = concentration of the target component in the feed before flotation (or cleaning);
# T = concentration of the target component in the tailings after flotation (or cleaning).

# Using these variables from the training data:
# C = rougher.output.concentrate_au
# F = rougher.input.feed_au
# T = rougher.output.tail_au

# Calculate the recovery for the 'rougher.output.recovery' feature
C = df_train_filled['rougher.output.concentrate_au']
F = df_train_filled['rougher.input.feed_au']
T = df_train_filled['rougher.output.tail_au']

# Perform the given formula to calculate recovery
recovery_calculated = (C * (F - T)) / ((F * (C - T))) * 100

# Drop NA from calculated recovery and original for accurate comparison
recovery_calculated_clean = recovery_calculated.dropna()
recovery_original = df_train_filled['rougher.output.recovery'].dropna()

# Align indices to compare the same entries
recovery_calculated_aligned = recovery_calculated_clean.loc[recovery_original.index]

# Calculate MAE between the calculated recovery and the feature values
mae = mean_absolute_error(recovery_original, recovery_calculated_aligned)
mae
```

```python
# Define the sMAPE function
def smape(actual, forecast):
    """
    Calculate the symmetric Mean Absolute Percentage Error (sMAPE) between two series.
    """
    actual = np.array(actual)
    forecast = np.array(forecast)
    denominator = np.abs(actual) + np.abs(forecast)
    safe_denominator = np.where(denominator == 0, 1, denominator)  # prevent division by zero
    diff = (np.abs(actual - forecast) / (safe_denominator / 2)) * 100
    return np.mean(diff)
    
    
    
    
    
# Define the final sMAPE function
def final_smape(smape_rougher, smape_final):
    """
    Calculate the final sMAPE value as the average of the sMAPE for the rougher and final stages.
    """
    return 0.25 * smape_rougher + 0.75 * smape_final    


train_predictions_rougher = df_train_filled['rougher.output.recovery'] + np.random.normal(0, 10, len(df_train))
train_predictions_final = df_train_filled['final.output.recovery'] + np.random.normal(0, 10, len(df_train))

smape_rougher = smape(df_train_filled['rougher.output.recovery'], train_predictions_rougher)
smape_final = smape(df_train_filled['final.output.recovery'], train_predictions_final)

# Calculate final sMAPE using the provided final_smape function
final_smape_value = final_smape(smape_rougher, smape_final)

print("Rougher sMAPE:", smape_rougher)
print("Final sMAPE:", smape_final)
print("Final sMAPE Value:", final_smape_value)




```
