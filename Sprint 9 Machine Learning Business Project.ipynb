```python
# Function to train model and predict
def train_and_predict(data):
    X = data[['f0', 'f1', 'f2']]
    y = data['product']
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25, random_state=42)
    model = LinearRegression()
    model.fit(X_train, y_train)
    predictions = model.predict(X_val)
#     return predictions, y_val, model
    # Make predictions
    y_pred = model.predict(X_val)
    
    # 2.3 Save predictions and correct answers
    results = pd.DataFrame({'Actual': y_val, 'Predicted': y_pred})
 
    # 2.4 Compute metrics
    average_predicted = y_pred.mean()
    rmse = mean_squared_error(y_val, y_pred, squared=False)
    
    return results, average_predicted, rmse

# Apply the function to each region's dataset
results_0, avg_pred_0, rmse_0 = train_and_predict(geo_data_0)
results_1, avg_pred_1, rmse_1 = train_and_predict(geo_data_1)
results_2, avg_pred_2, rmse_2 = train_and_predict(geo_data_2)

# Print the results
print("Region 0: Average Predicted Volume =", avg_pred_0, "RMSE =", rmse_0)
print("Region 1: Average Predicted Volume =", avg_pred_1, "RMSE =", rmse_1)
print("Region 2: Average Predicted Volume =", avg_pred_2, "RMSE =", rmse_2)

# 2.5 Analyze the results
# Comparing RMSE and average predicted volumes across regions
results_analysis ={
    "Region 0": {"Average Predicted": avg_pred_0, "RMSE": rmse_0},
    "Region 1": {"Average Predicted": avg_pred_1, "RMSE": rmse_1},
    "Region 2": {"Average Predicted": avg_pred_2, "RMSE": rmse_2}
}
```

```python
# Define the profit calculation function
def calculate_profit(data_val, number_of_wells, revenue_per_unit, cost_per_well):
    
    top_wells = data_val.nlargest(number_of_wells, 'Predicted')
    total_volume = top_wells['Actual'].sum()
    total_revenue = total_volume * revenue_per_unit
    total_cost = number_of_wells * cost_per_well
    profit = total_revenue - total_cost
    
    return profit, total_volume

# Calculate profit for each region
profit_0, volume_0 = calculate_profit(results_0, number_of_wells, revenue_per_unit, cost_per_well)
profit_1, volume_1 = calculate_profit(results_1, number_of_wells, revenue_per_unit, cost_per_well)
profit_2, volume_2 = calculate_profit(results_2, number_of_wells, revenue_per_unit, cost_per_well)

# Suggest the best region based on maximum profit
profits = {'Region 0': profit_0, 'Region 1': profit_1, 'Region 2': profit_2}
best_region = max(profits, key=profits.get)

print(f"Best region for development: {best_region}")
print(f"Profit: ${profits[best_region]:,.2f}")
print(f"Total Volume of Reserves: {locals()['volume_' + best_region[-1]]} thousand barrels")
```

```python
# Function to perform bootstrapping
def bootstrap_profit(data, n_iterations, random_state=42):
    profits = []
    rs = np.random.RandomState(random_state)
#     data = pd.DataFrame({'predictions': predictions, 'actual': actual})
    for _ in range(n_iterations):
        sample = data.sample(500, replace=True, random_state=rs)
#         top_wells = data.loc[sample.index]
        profit, volume = calculate_profit(sample, number_of_wells, revenue_per_unit, cost_per_well)
        profits.append(profit)

    profits = np.array(profits)
    mean_profit = np.mean(profits)
    confidence_interval = np.percentile(profits, [2.5, 97.5])
    risk_of_loss = (profits < 0).mean() * 100
    return mean_profit, confidence_interval, risk_of_loss

# Perform bootstrapping for each region
mean_profit_0, ci_0, risk_0 = bootstrap_profit(results_0, 1000, random_state=42)
mean_profit_1, ci_1, risk_1 = bootstrap_profit(results_1, 1000, random_state=42)
mean_profit_2, ci_2, risk_2 = bootstrap_profit(results_2, 1000, random_state=42)

# Summarize results
results = {
    'Region 0': {'Average Profit': mean_profit_0, '95% CI': ci_0, 'Risk of Loss (%)': risk_0},
    'Region 1': {'Average Profit': mean_profit_1, '95% CI': ci_1, 'Risk of Loss (%)': risk_1},
    'Region 2': {'Average Profit': mean_profit_2, '95% CI': ci_2, 'Risk of Loss (%)': risk_2}
}

# Find the best region based on average profit and acceptable risk
best_region = min(results, key=lambda x: (results[x]['Risk of Loss (%)'] > 2.5, -results[x]['Average Profit']))

print("Analysis Results for Each Region:")
for region, info in results.items():
    print(f"{region}: Average Profit = ${info['Average Profit']:,.2f}, 95% CI = {info['95% CI']}, Risk of Loss = {info['Risk of Loss (%)']}%")

print(f"\nSuggested Region for Development: {best_region} based on highest average profit and acceptable risk.")
```
