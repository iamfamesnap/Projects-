````python
# Test the hypotheses
from scipy import stats as st
import numpy as np
import pandas as pd


alpha = 0.05 # critical statistical significance
results = st.ttest_ind(surf_plan_revenue, ultimate_plan_revenue)
print('p_value: ', results.pvalue)
if results.pvalue < alpha: 
    print('we reject the null hypothesis')
    
else: 
    print('we cant reject the null hypothesis')


(results.pvalue)
````

````python
# Test the hypothesis

# Split data into groups that include NY-NJ regions and others to compare the two. 
ny_nj_revenue = merged_df[merged_df['city'] == 'NY-NJ']['total_revenue']
other_revenue = merged_df[merged_df['city'] != 'NY-NJ']['total_revenue']

# Perform two_sample t-test 
alpha = 0.05
reesults = st.ttest_ind(ny_nj_revenue, other_revenue)
print('p_value: ', results.pvalue)
if results.pvalue < alpha:
    print('we reject the null hypothesis. The average revenue from NY-NJ users is significantly diffrent from that of other regions.')
    
else: 
    print('We cant reject the null hypothesis. There is no significant difference in the average revenue between NY-NJ and other regions.')
````
