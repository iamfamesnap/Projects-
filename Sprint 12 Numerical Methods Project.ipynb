````python
# Encode properly after splitting

categorical_cols = ['VehicleType', 'Gearbox', 'FuelType', 'NotRepaired']

# Make copies so we don't overwrite
X_train_encoded = X_train.copy()
X_val_encoded = X_val.copy()
X_test_encoded = X_test.copy()

# Apply LabelEncoder separately for each column based on training data

for col in categorical_cols:
    lencoder = LabelEncoder()
    X_train_encoded[col] = lencoder.fit_transform(X_train_encoded[col])
    
# Use same encoder for val/test to prevent mismatch
    X_val_encoded[col] = lencoder.transform(X_val_encoded[col])
    X_test_encoded[col] = lencoder.transform(X_test_encoded[col])
    
````
